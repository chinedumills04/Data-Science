```python
import numpy as np
import time

def timeit(somefunc, *args, repeats=10, **kwargs):
    times = []

    while repeats > 0:
        starttime = time.time()
        ans = somefunc(*args, **kwargs)
        endtime = time.time()

        timetaken = endtime - starttime
        times.append(timetaken)
        repeats -= 1

        print(f"Running Time: {timetaken} seconds")

    mean = np.mean(times)
    stdev = np.std(times)

    return mean, stdev
```

```python
import timeit
import numpy as np

def jaccard_similarity(doc1, doc2):
    set1 = set(doc1)
    set2 = set(doc2)
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union

def cosine_similarity(doc1, doc2):
    dot_product = np.dot(doc1, doc2)
    norm_A = np.linalg.norm(doc1)
    norm_B = np.linalg.norm(doc2)
    return dot_product / (norm_A * norm_B)

def time_it(func, *args, number_of_repeats=1):
    return timeit.timeit(lambda: func(*args), number=number_of_repeats)

def estimate_constant_similarity(similarity_function, documents, number_of_repeats=100):
    num_docs = len(documents)
    total_time = 0

    for i in range(num_docs):
        for j in range(num_docs):
            if i != j:  # Avoid self-comparison
                start_time = timeit.default_timer()
                similarity_function(documents[i], documents[j])
                end_time = timeit.default_timer()
                total_time += (end_time - start_time)

    average_time = total_time / (num_docs * (num_docs - 1))
    constant_estimate = average_time / (num_docs * (num_docs - 1))

    return constant_estimate

# Example: Generate random data for testing
num_docs = 20
doc_size = 100
documents = [np.random.choice(range(10), size=doc_size, replace=True) for _ in range(num_docs)]

# Estimate the constant for Jaccard similarity
jaccard_constant = estimate_constant_similarity(jaccard_similarity, documents)
print(f"Estimated constant for Jaccard similarity: {jaccard_constant}")

# Estimate the constant for Cosine similarity
cosine_constant = estimate_constant_similarity(cosine_similarity, documents)
print(f"Estimated constant for Cosine similarity: {cosine_constant}")
```

```python
import numpy as np

# Cosine function is defined here
def cosine(doc1, doc2):
    dot_product = np.dot(doc1, doc2)
    magnitude_v1 = np.linalg.norm(doc1)
    magnitude_v2 = np.linalg.norm(doc2)
    
    if magnitude_v1 == 0 or magnitude_v2 == 0:
        return 0  # Avoid division by zero
    
    return dot_product / (magnitude_v1 * magnitude_v2)

# Creating a list of document vectors
doc_size = 100
num_docs = 20
docvectors = [np.random.rand(doc_size) for _ in range(num_docs)]

# Code for calculating similarities
sims = np.zeros((num_docs, num_docs))  # Initialize a matrix to store similarities

for i in range(num_docs):
    for j in range(i + 1, num_docs):  # Only calculate for j > i to avoid redundant calculations
        similarity = cosine(docvectors[i], docvectors[j])
        sims[i][j] = similarity
        sims[j][i] = similarity  # Fill in the symmetric entry

print(sims)

maxsim = -1
minsim = 1
maxi = -1
maxj = -1
mini = -1
minj = -1

for i in range(num_docs):
    for j in range(i + 1, num_docs):
        if sims[i][j] > maxsim:
            maxsim = sims[i][j]
            maxi = i
            maxj = j
        if sims[i][j] < minsim:
            minsim = sims[i][j]
            mini = i
            minj = j

print("Minimum similarity of {} at {},{}".format(minsim, mini, minj))
print("Maximum similarity of {} at {},{}".format(maxsim, maxi, maxj))
```
